```python
def string_clean_capital_fn(dirtyStr):
    """ remove whitespaces and clean strings """

    str1 = dirtyStr.replace('_', ' ')
    str2 = str1.replace('-', ' ')
    str3 = str2.capitalize()
    cleanStr = str3.strip()
    return (cleanStr)
```

```python
def extract_grass_botanical_fn(df, index, row, string_clean_capital_fn, n):
    list_botanical = []
    species1 = (n + '_SP:' + n + '1')
    botanical1 = string_clean_capital_fn(str(row[species1]))
    print(botanical1)

    if botanical1 != 'Nan' and botanical1 != 'Other1':
        list_botanical.append(botanical1)
    elif botanical1 == 'Other1':
        botanical1 = string_clean_capital_fn(str(row[n + '_SP:' + n + '_OTHER1']))
    else:
        botanical1 = np.nan

    botanical2 = string_clean_capital_fn(str(row[n + '_SP:' + n + '2']))

    if botanical2 != 'Nan' and botanical2 != 'Other2':
        list_botanical.append(botanical2)
    elif botanical2 == 'Other2':
        botanical2 = string_clean_capital_fn(str(row[n + '_SP:' + n + '_OTHER2']))
    else:
        botanical2 = np.nan

    botanical3 = string_clean_capital_fn(str(row[n + '_SP:' + n + '3']))

    if botanical3 != 'Nan' and botanical3 != 'Other3':
        list_botanical.append(botanical3)
    elif botanical3 == 'Other3':
        botanical3 = string_clean_capital_fn(str(row[n + '_SP:' + n + '_OTHER3']))
    else:
        botanical3 = np.nan

    botanical4 = string_clean_capital_fn(str(row[n + '_SP:' + n + '4']))

    if botanical4 != 'Nan' and botanical4 != 'Other4':
        list_botanical.append(botanical4)
    elif botanical4 == 'Other4':
        botanical4 = string_clean_capital_fn(str(row[n + '_SP:' + n + '_OTHER4']))
    else:
        botanical4 = np.nan
    print(botanical1, botanical2, botanical3, botanical4)
    return botanical1, botanical2, botanical3, botanical4
```
